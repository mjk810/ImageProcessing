```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('enzyme.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap = 'gray')
```

```python
_, img_binary = cv2.threshold(img_gray, 50, 255, cv2.THRESH_BINARY)
plt.imshow(img_binary, cmap='gray')
```

```python
_, con, _=cv2.findContours(img_binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
con_img = img.copy()
contoured_img = cv2.drawContours(con_img, con, -1, (0,255,0), 2)
plt.imshow(contoured_img)
```

```python
plt.imshow(img)
```

```python
#add bounding rectangle
x,y,w,h = cv2.boundingRect(con[0])
box_img = cv2.rectangle(img, (x,y), (x+w, y+h), (200, 0, 200), 2)
plt.imshow(box_img)
```

```python
#crop to box
cropped_img = box_img.copy()
cropped_img = cropped_img[y:y+h, x:x+w]
plt.imshow(cropped_img)
```
