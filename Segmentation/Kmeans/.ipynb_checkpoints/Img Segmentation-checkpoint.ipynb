```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('farm1.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#reshape the image and convert to float
img_reshape = img_rgb.reshape((-1,3))
img_reshape = np.float32(img_reshape)
```

```python
#implement k means
k=3
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_COUNT, 10, 1.0)
retval, labels, centers = cv2.kmeans(img_reshape, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
centers = np.uint8(centers)
seg_data = centers[labels.flatten()]
seg_img = seg_data.reshape(img.shape)
labels_reshape = labels.reshape((img.shape[0], img.shape[1]))
```

```python
plt.imshow(seg_img)
```

```python
#mask the img by cluster
cluster_number = 1
masked_img = np.copy(img_rgb)
masked_img[labels_reshape==cluster_number]=[255,0,255]
```

```python
plt.imshow(masked_img)
```

```python
#visualize clusters
cluster_one = 0
plt.imshow(labels_reshape==cluster_one)
```

```python
cluster_two = 1
plt.imshow(labels_reshape==cluster_two)
```
