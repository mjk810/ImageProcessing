```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
filename = 'transformerBlue.jpeg'
img = cv2.imread(filename, 1)
print('shape: ', img.shape)
plt.imshow(img)
```

```python
#change from bgr to rgb
img_copy = np.copy(img)
rgb_image = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
#crop black bars (visual estimation)
rgb_image_cropped = rgb_image[0:828,200:1600]
plt.imshow(rgb_image_cropped)
```

```python
#create a mask to separate the transformer from the background using cv2 inrange function
copy_cropped_image = rgb_image_cropped.copy()
lower_blue = np.array([0,0,160])
upper_blue = np.array([20,190,255])
mask = cv2.inRange(copy_cropped_image, lower_blue, upper_blue)

plt.imshow(mask, cmap = 'gray')
```

```python
masked_img = rgb_image_cropped.copy()
masked_img[mask!=0]=[0,0,0]
plt.imshow(masked_img)
print('image size', masked_img.shape)
```

```python
#get the new background
background_image = cv2.imread('vangogh.jpg')
#convert to rgb
rgb_background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)
#crop to size
rgb_background_image = rgb_background_image[0:828, 0:1400]
plt.imshow(rgb_background_image)
```

```python
#cut out shape from background
masked_background = rgb_background_image.copy()
masked_background[mask==0]=[0,0,0]
plt.imshow(masked_background)
```

```python
#add background and object images together to produce final image
final_image = masked_background + masked_img
plt.imshow(final_image)
```
